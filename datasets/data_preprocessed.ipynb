```python
import numpy as np 
import torch.nn.functional as F
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

def MAE(score, target):
    if len(score.shape) == 1:
        mae = np.sum(np.abs(score.cpu().data.numpy() - target.cpu().data.numpy())) / len(target)
    # _, pred = score.max(1)
    # l1loss = nn.L1Loss()
    # target = target.float()
    # mae = l1loss(pred, target)
    else:
        s_dim, out_dim = score.shape
        probs = F.softmax(score, -1)
        probs_data = probs.cpu().data.numpy()
        target_data = target.cpu().data.numpy()
        max_data = np.argmax(probs_data, axis=1)
        label_arr = np.array(range(out_dim))
        exp_data = np.sum(probs_data * label_arr, axis=1)
        mae = sum(abs(exp_data - target_data)) * 1.0 / len(target_data)
    return mae

def specific_eval(score, target):
    # _, pred = score.max(1)
    prediction = score.cpu().data.numpy()
    true_lable = target.cpu().data.numpy()
    measure_result = classification_report(true_lable, prediction)
    measure_dict = classification_report(true_lable, prediction, output_dict=True)
    return measure_dict, measure_result

def metrics(score, target, dataset):
    if dataset=='Adience':
        hist = confusion_matrix(target, score, labels=[0, 1, 2, 3, 4, 5, 6, 7])
    else:
        hist = confusion_matrix(target, score, labels=[0, 1, 2, 3, 4])
    mae = MAE(score, target)

    measure_dict, measure_result = specific_eval(score, target)
    weight_f1 = measure_dict["weighted avg"]["f1-score"]
    weight_recall = measure_dict["weighted avg"]["recall"]
    weight_precision = measure_dict["weighted avg"]["precision"]
    macro_f1 = measure_dict["macro avg"]["f1-score"]
    macro_recall = measure_dict["macro avg"]["recall"]
    macro_precision = measure_dict["macro avg"]["precision"]

    OA = np.diag(hist).sum() / hist.sum()

    mAcc = np.diag(hist) / hist.sum(axis=1)
    mAcc = np.nanmean(mAcc)
    return OA, mAcc, mae, weight_f1, weight_recall, weight_precision, macro_f1, macro_recall, macro_precision
```
